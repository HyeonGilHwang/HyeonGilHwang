```python
from selenium import webdriver
import time

driver = webdriver.Chrome(r"C:\Users\tmxhs\0_crawiling\chromedriver.exe")

driver.implicitly_wait(3)

driver.get("https://www.op.gg/champion/statistics")

top_name = []
jungle_name = []
mid_nmae = []
ad_name = []
spt_name = []

for j in range(1, 6):
    driver.find_element_by_xpath('/html/body/div[2]/div[3]/div[2]/div[1]/div[6]/div/div[2]/div/div[1]/div/ul/li['+str(j)+']/a').click()
    time.sleep(2)
    for i in range(1, 6):
        a = driver.find_element_by_xpath('/html/body/div[2]/div[3]/div[2]/div[1]/div[6]/div/div[2]/div/div[1]/div/table/tbody['+str(j)+']/tr['+str(i)+']/td[4]/a/div[1]')
        if j == 1:
            top_name.append(a.text)
        elif j == 2:
            jungle_name.append(a.text)
        elif j == 3:
            mid_nmae.append(a.text)
        elif j == 4:
            ad_name.append(a.text)
        else:
            spt_name.append(a.text)
            
print(top_name)
print(jungle_name)
print(mid_nmae)
print(ad_name)
print(spt_name)

time.sleep(5)
driver.close()
```
